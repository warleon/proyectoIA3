```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
dbscan = pd.read_csv("results/dbscan.csv")
kmeans = pd.read_csv("results/kmeans.csv")
```

```python
for (algo,name) in [(dbscan,"dbscan"),(kmeans,"kmeans")]:
    plt.title(name)
    plt.xlabel("Random Attribute")
    plt.ylabel("Random Attribute")
    for val in algo.cluster.unique():
        plt.plot(algo[algo.cluster == val]["dim1"],algo[algo.cluster == val]["dim2"],"*")
    plt.savefig("results/"+name+".jpg")
    plt.show()
```
